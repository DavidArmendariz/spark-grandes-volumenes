```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n = 500
```

```python
x11 = np.random.normal(100, 15, n)
x12 = np.random.normal(-20, 15, n)
x13 = np.random.normal(100, 15, n)
X1 = np.concatenate([x11, x12, x13])
x21 = np.random.normal(-20, 15, n)
x22 = np.random.normal(10, 20, n)
x23 = np.random.normal(40, 20, n)
X2 = np.concatenate([x21, x22, x23])

df = pd.DataFrame({"x1": X1, "x2": X2})

df.to_csv("data.csv", index=False, header=True)
```

```python
plt.scatter(X1, X2)
plt.title("Plot of the raw data")
plt.show()
```

```python
plt.scatter(x11, x21, c="r")
plt.scatter(x12, x22, c="g")
plt.scatter(x13, x23, c="b")
plt.title("Plot of the clusters")
plt.show()
```
